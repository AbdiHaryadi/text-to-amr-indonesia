```python
file_path = "amr_simple_dev.txt"
```

```python
with open(file_path) as f:
  data = f.readlines()
```

```python
data[:12]
```

```python
list_amr = []
list_sent = []

sent_now = ""
amr_now = ""
is_reading_amr = False

for line in data:
  line = line.strip()

  # print(line)
  if (line==''):
    if (is_reading_amr):
      is_reading_amr = False
      amr_now = " ".join(amr_now.split())
      list_amr.append(amr_now)
      amr_now = ""
    continue

  line += "."

  if (line[0]=='#'):
    if ('# ::snt' in line):
      sent_now = line[8:-1]
      list_sent.append(sent_now)
  else:  #reading amr
    is_reading_amr = True
    found_slash = False # /
    found_colon = False # :
    for c in line:
      if (c=='.'):
        amr_now += " "
        found_colon = False
        found_slash = False
      elif (c=='('):
        amr_now += " " + c + " "
      elif (c==')' and not found_colon and not found_slash):
        amr_now += " " + c + " "
      elif (c=='/'):
        found_slash = True
      elif (c==':'):
        found_colon = True
        amr_now += c
      else:
        if (found_colon):
          amr_now += c
          if (c==' '):
            found_colon = False
        elif (found_slash):
          if (c==')'):
          
            amr_now += " "
            found_slash = False
          amr_now += c
amr_now = " ".join(amr_now.split())
list_amr.append(amr_now)
            
```

```python
print(len(list_amr))
print(len(list_sent))
```

```python
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')

word_tokenize("halo nama saya taufiq, hahaha")
```

```python
list_tokenized_sent = []
for sent in list_sent:
  list_tokenized_sent.append(" ".join(word_tokenize(sent)))
```

```python
list_tokenized_sent[:20]
```

```python
f = open("dev.source", "w")
for amr in list_amr:
  f.write(amr)
  f.write("\n")
f.close()

f = open("dev.target", "w")
for sent in list_sent:
  f.write(sent)
  f.write("\n")
f.close()

f = open("dev.target.tok", "w")
for sent in list_tokenized_sent:
  f.write(sent)
  f.write("\n")
f.close()
```
